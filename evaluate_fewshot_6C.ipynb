```python
import pandas as pd

#csv_file_onek_GLASSO = 'evaluation/fewshots/6C_paddle_woverlap_False_adaptatif_1000%_GLASSO_analysis.csv'
csv_file_overlap_onek_GLASSO = 'evaluation/fewshots/6C_paddle_woverlap_True_adaptatif_1000%_GLASSO_analysis.csv'
csv_file_overlap_fivek_GLASSO = 'evaluation/fewshots/6C_paddle_woverlap_True_adaptatif_5000%_GLASSO_analysis.csv'
csv_file_overlap_paper_GLASSO = 'evaluation/fewshots/6C_paddle_woverlap_True_1250_GLASSO_analysis.csv'
#csv_file_onek_NONE = 'evaluation/fewshots/6C_paddle_woverlap_False_adaptatif_1000%_NONE_analysis.csv'
csv_file_overlap_onek_NONE = 'evaluation/fewshots/6C_paddle_woverlap_True_adaptatif_1000%_NONE_analysis.csv'
csv_file_overlap_fivek_NONE = 'evaluation/fewshots/6C_paddle_woverlap_True_adaptatif_5000%_NONE_analysis.csv'
csv_file_overlap_paper_NONE = 'evaluation/fewshots/6C_paddle_woverlap_True_1250_NONE_analysis.csv'


#df_onek_GLASSO = pd.read_csv(csv_file_onek_GLASSO)
df_overlap_onek_GLASSO = pd.read_csv(csv_file_overlap_onek_GLASSO)
#df_onek_NONE = pd.read_csv(csv_file_onek_NONE)
df_overlap_fivek_GLASSO = pd.read_csv(csv_file_overlap_fivek_GLASSO)
df_overlap_paper_GLASSO = pd.read_csv(csv_file_overlap_paper_GLASSO)

df_overlap_onek_NONE = pd.read_csv(csv_file_overlap_onek_NONE)
df_overlap_fivek_NONE = pd.read_csv(csv_file_overlap_fivek_NONE)
df_overlap_paper_NONE = pd.read_csv(csv_file_overlap_paper_NONE)

```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
plt.figure(figsize=(13, 8))
#plt.plot(df_onek_GLASSO['n_shots'], df_onek_GLASSO['accuracies'], marker='o', linestyle='-', color='b', label='PADDLE-Cov alpha=1000%')
plt.plot(df_overlap_onek_GLASSO['n_shots'], df_overlap_onek_GLASSO['accuracies'], marker='o', linestyle='-', color='r', label='PADDLE-Cov alpha=1000% w overlap')
#plt.plot(df_onek_NONE['n_shots'], df_onek_NONE['accuracies'], marker='o', linestyle='-', color='g', label='PADDLE alpha=1000%')
#plt.plot(df_overlap_onek_NONE['n_shots'], df_overlap_onek_NONE['accuracies'], marker='o', linestyle='-', color='y', label='PADDLE alpha=1000% w overlap')
plt.plot(df_overlap_fivek_GLASSO['n_shots'], df_overlap_fivek_GLASSO['accuracies'], marker='o', linestyle='-', color='c', label='PADDLE-Cov alpha=5000% w overlap')
plt.plot(df_overlap_paper_GLASSO['n_shots'], df_overlap_paper_GLASSO['accuracies'], marker='o', linestyle='-', color='m', label='PADDLE-Cov alpha=1250 w overlap')

plt.xscale('log')
plt.xlabel('n_shots')
plt.ylabel('accuracy')
plt.title('Fewshots accuracies analysis')
plt.legend()
#plt.savefig('evaluation/fewshots/6C_paddle_accuracy_analysis.png')
plt.show()
```
