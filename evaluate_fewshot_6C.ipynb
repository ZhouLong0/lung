```python
import pandas as pd

csv_file = 'evaluation/fewshots/6C_paddle_n_shots_analysis.csv'

df = pd.read_csv(csv_file)

df
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 6))
plt.plot(df['n_shots'], df['accuracy'], marker='o', linestyle='-', color='b', label='Accuracy')
plt.xlabel('n_shots')
plt.ylabel('accuracy')
plt.title('Fewshots analysis')
plt.savefig('evaluation/fewshots/6C_paddle_n_shots_analysis.png')
plt.show()
```
