```python
#config
from src.utils.LungDataset import LungSet
import pandas as pd

query_file_name = "7C_query"
support_file_name = "7C_support_augmented"

split_path = "data/splits/" 

query_data_folder = "data/normalized_data"
support_data_folder = "data/augmented_data"

df_query = pd.read_csv(split_path + query_file_name + '.csv').sample(frac = 1)
df_support = pd.read_csv(split_path + support_file_name + '.csv').sample(frac = 1)

query_dataset = LungSet(df_query, data_path=query_data_folder, file_name=True)
support_dataset = LungSet(df_support, data_path=support_data_folder, file_name=True)
```

```python
exctract_folder = 'checkpoints/fewshot/resnet18/extracted_features/'

query_features_dict = fs_utils.extract_features(model, dataset=query_dataset, save_directory = exctract_folder, save_filename = query_file_name + '.plk')

support_fetures_dict = fs_utils.extract_features(model, dataset=support_dataset, save_directory = exctract_folder, save_filename = support_file_name + '.plk')
```
