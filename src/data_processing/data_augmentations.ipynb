```python
from PIL import Image, ImageOps
import numpy as np
import random
from skimage.transform import rotate
import matplotlib.pyplot as plt
```

```python
def horizontal_flip(image):
    return ImageOps.mirror(image)

def vertical_flip(image):
    return ImageOps.flip(image)

def rotation(image, degree):
    return rotate(np.array(image), degree, mode="reflect")
```

```python
# Example usage
input_image_path = "/home/aymen/Desktop/remote_jensen/AymenRouen/lung_cancer_Rouen/dataset/data/Acinaire/2_B_row_4_col_6.jpg"
original_image = Image.open(input_image_path)

# Apply transformations
flipped_horizontally = horizontal_flip(original_image)
flipped_vertically = vertical_flip(original_image)
rotated0 = rotation(original_image, 45)
rotated1 = rotation(original_image, 90)
rotated2 = rotation(original_image, 135)
rotated3 = rotation(original_image, 225)
rotated4 = rotation(original_image, 270)
rotated5 = rotation(original_image, 315)

fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(12, 12))
ax=axes.flatten()
i=0
titles = ["Original", "Horizontally Flipped", "Vertically Flipped", "Rotated 45", "Rotated 90", "Rotated 135", "Rotated 225", "Rotated 270", "Rotated 315"]
for t in [original_image, flipped_horizontally, flipped_vertically, rotated0, rotated1, rotated2, rotated3, rotated4, rotated5]:
    ax[i].imshow(t)
    ax[i].set_title(titles[i])
    ax[i].axis('off')
    i+=1
plt.tight_layout()
plt.show()
```
